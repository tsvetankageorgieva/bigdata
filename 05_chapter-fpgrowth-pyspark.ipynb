```python
import pyspark.sql, pyspark.sql.functions
from pyspark.sql import SparkSession

# Стартиране на SparkSession 
spark = (
    SparkSession.builder
    .appName("FP-growth-example")
    .master("local[*]")
    .getOrCreate()
)
print("SparkSession стартиран успешно")
print("Spark версия:", spark.version)
```

```python
from pyspark.ml.fpm import FPGrowth

# Примерни транзакции
data = [(["Milk", "Bread", "Eggs"],),
    (["Milk", "Butter"],),
    (["Bread", "Butter", "Beer"],),
    (["Milk", "Bread", "Butter"],),
    (["Bread", "Beer"],)]

# Създаване на DataFrame
df = spark.createDataFrame(data, ["items"])

# Създаване на модел
fpGrowth = FPGrowth(itemsCol="items",minSupport=0.4,minConfidence=0.7)
# Обучение на модела
model = fpGrowth.fit(df)
# Често срещани набори от обекти
frequent_itemsets = model.freqItemsets
# Асоциативни правила
association_rules = model.associationRules

# Извеждане на резултатите
print("Често срещани набори от обекти:")
frequent_itemsets.show()
print("Асоциативни правила:")
association_rules.show()
```
