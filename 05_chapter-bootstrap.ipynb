```python
from sklearn.utils import resample
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np

# Зареждане на данните
df = pd.read_csv('students-online.csv')
# Входни характеристики
X = df[['Age', 'PlatformHours', 'OutPlatformHours']]
# Целева характеристика
y = df['Satisfaction']

# Списък за резултатите
mse_scores = []
# Брой bootstrap итерации
n_iterations = 100
for i in range(n_iterations):
    # Bootstrap извадка
    X_boot, y_boot = resample(X,y,replace=True,random_state=i)

    # Създаване на модел
    model = LinearRegression()
    # Обучение
    model.fit(X_boot, y_boot)
    # Прогнозиране върху оригиналните данни
    y_pred = model.predict(X)

    # Изчисляване на MSE
    mse = mean_squared_error(y, y_pred)
    mse_scores.append(mse)

# Средна стойност
print("Средна MSE:")
print(np.mean(mse_scores))
```
