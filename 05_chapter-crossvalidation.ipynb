```python
from sklearn.model_selection import cross_val_score, StratifiedKFold, LeaveOneOut
from sklearn.linear_model import LinearRegression
import pandas as pd

# Зареждане на данните
df = pd.read_csv('students-online.csv')

# Входни характеристики
X = df[['Age', 'PlatformHours', 'OutPlatformHours']]
# Целева характеристика
y = df['Satisfaction']

# Създаване на модел
model = LinearRegression()
# Кръстосана валидация с отделяне на един обект
cv = LeaveOneOut()
# 5-кратна cross-validation
scores = cross_val_score(model,X,y,cv=cv,scoring='neg_mean_squared_error')
# Преобразуване в положителни стойности
mse_scores = -scores

#cv = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
print("MSE за отделните разделяния:")
print(mse_scores)

print("\nСредна MSE:")
print(mse_scores.mean())
```
