```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import pandas as pd
from lime.lime_tabular import LimeTabularExplainer

# Зареждане на данните
data = pd.read_csv('students-online.csv')
# Избор само на числовите колони
numeric_df = data.select_dtypes(include=['number'])

# Целева променлива
y = numeric_df["Satisfaction"]
# Входни характеристики
X = numeric_df.drop(columns=["Satisfaction", "ID"])
# Разделяне на данните
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)

# Създаване на невронна мрежа
model = MLPClassifier(hidden_layer_sizes=(10, 5),
    activation='relu', max_iter=1000, random_state=42)
# Обучение
model.fit(X_train, y_train)

# Предварително зададени имена на характеристиките
feature_names = ["Age","PlatformHours","OutPlatformHours","Self-assessment"]
# Предварително зададени класове
class_names = ["1", "2", "3", "4", "5"]
# Създаване на LIME обяснител
explainer = LimeTabularExplainer(
    training_data=X_train.values,
    feature_names=feature_names,
    class_names=class_names,
    mode='classification')
# Избор на запис за обяснение
instance = X_test.iloc[2].values
# Генериране на обяснение
explanation = explainer.explain_instance(
    instance, model.predict_proba,
    num_features=4)

# Извеждане на обяснението
print(explanation.as_list())
```
