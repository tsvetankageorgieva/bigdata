```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
import pandas as pd

# Данни
df = pd.read_csv('students-online.csv')
# Целева характеристика
y = df['Satisfaction']
# Входни характеристики
X = df[['Age','PlatformHours','OutPlatformHours','Self-assessment']]

# Избор на 2 най-информативни характеристики
selector = SelectKBest(score_func=chi2, k=2)
# Обучение и трансформация
X_new = selector.fit_transform(X, y)
# Индекси на избраните характеристики
selected_features = selector.get_support(indices=True)

print("Избрани характеристики:")
for index in selected_features:
    print(X.columns[index])
```
