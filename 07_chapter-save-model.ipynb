```python
from sklearn.linear_model import LogisticRegression
import pickle

# 1. Данни
X = [[0], [1], [2], [3]]
y = [0, 0, 1, 1]

# 2. Обучение на модел
model = LogisticRegression()
model.fit(X, y)

# 3. Записване на модела във файл (.pkl)
pickle.dump(model, open("model_reg.pkl", "wb") )
```

```python
from sklearn.neighbors import KNeighborsClassifier
import pickle

# 1. Данни
X = [[0], [1], [2], [3]]
y = [0, 0, 1, 1]

# 2. Обучение на модел
model = KNeighborsClassifier(n_neighbors = 1)
model.fit(X, y)

# 3. Записване на модела във файл (.pkl)
pickle.dump(model, open("model_knn.pkl", "wb") )
```

```python
import pickle

model = pickle.load(open("model_knn.pkl", "rb") )

# използване на модела
print(model.predict([[1.5]]))
```

```python
from sklearn.linear_model import LogisticRegression
from joblib import dump, load

# 1. Данни
X = [[0], [1], [2], [3]]
y = [0, 0, 1, 1]

# 2. Обучение на модел
model = LogisticRegression()
model.fit(X, y)

# 3. Записване на модела във файл .joblib
dump(model, "model.joblib")
```

```python
from joblib import load

model = load("model.joblib")

print(model.predict([[1.5]]))
```
