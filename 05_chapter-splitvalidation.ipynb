```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

# Зареждане на данните
df = pd.read_csv('students-online.csv')
# Входни характеристики
X = df[['Age', 'PlatformHours', 'OutPlatformHours']]
# Целева характеристика
y = df['Satisfaction']

# Разделяне на данните: 70% обучаващ набор и 30% тестов набор
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Създаване на модел
model = LinearRegression()
# Обучение
model.fit(X_train, y_train)
# Прогнозиране
y_pred = model.predict(X_test)

# Оценка на модела
mse = mean_squared_error(y_test, y_pred)

print("Средноквадратична грешка:", mse)
```
