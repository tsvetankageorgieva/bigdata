```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Зареждане на примерни данни
iris = load_iris()

X = iris.data
y = iris.target

# Разделяне на обучаващ и тестов набор
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Създаване на невронна мрежа
model = MLPClassifier(hidden_layer_sizes=(10, 5),  # два скрити слоя
    activation='relu', alpha=0.01, 
    max_iter=1000, solver='adam',
    random_state=42)

# Обучение на модела
model.fit(X_train, y_train)

# Прогнозиране
y_pred = model.predict(X_test)

# Оценка на модела
accuracy = accuracy_score(y_test, y_pred)
print("Точност:", accuracy)
```
