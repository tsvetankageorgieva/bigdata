```python
import pandas as pd

# Път към файла (ако е в същата папка)
file_path = "customers-data.csv"

# Зареждане на CSV файла
df = pd.read_csv(file_path)

# Показване на редовете
df

# Информация за типовете данни
# print(df.info())
```

```python
# Премахване на редове с липсващи стойности в name

df_clean = df.dropna(subset=["name"])
df_clean
```

```python
# Попълване на липсващи стойности 
# Средна стойност за purchase_amount, медиана за age, Unknown за name
df_filled = df.copy()

df_filled["purchase_amount"] = df_filled["purchase_amount"].fillna(df_filled["purchase_amount"].mean())
df_filled["age"] = df_filled["age"].fillna(df_filled["age"].median())
df_filled["name"] = df_filled["name"].fillna("Unknown")

print(df_filled)
```

```python
# Добавяне на индикаторна колона за липсващи стойности в purchase_date
df_filled["missing_purchase_date_flag"] = df_filled["purchase_date"].isna().astype(int)
df_filled
```

```python
# Откриване на дубликати
df_filled[df_filled.duplicated()]

# Премахване на дубликати
df_no_duplicates = df_filled.drop_duplicates()
df_no_duplicates
```

```python
# Откриване на отрицателни възрасти или над 120
invalid_age = df[(df["age"] < 0) | (df["age"] > 120)]
invalid_age

# Замяна на отрицателните възрасти и възрасти над 120 с NaN
df_corrected = df.copy()
df_corrected["age"] = df["age"].apply(lambda x: None if (x < 0 or x > 120) else x)
df_corrected
```

```python
# Преобразуване на колоната в тип datetime
df['purchase_date'] = pd.to_datetime(df['purchase_date'])

# Създаване на нови колони: ден от седмицата и месец
df['day_of_week'] = df['purchase_date'].dt.day_name()
df['month'] = df['purchase_date'].dt.month_name()

print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

# Кодиране с етикети (label encoding) – всяка категория се заменя с цяло число
encoder = LabelEncoder()
df["category_encoded"] = encoder.fit_transform(df["product_category"])
print(df)
```

```python
# кодиране с бинарни стойности (one-hot encoding) за product_category
df_ohe = pd.get_dummies(df, columns=["product_category"], prefix="cat")
print(df_ohe)
```

```python
# Min–Max скалиране
from sklearn.preprocessing import MinMaxScaler

# Избираме числовите колони за скалиране
numeric_cols = ["age", "purchase_amount"]

# Създаваме MinMaxScaler
scaler = MinMaxScaler()

# Прилагаме скалирането (пропуска None/NaN стойности)
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print(df)
```

```python
import numpy as np

# Логаритмична трансформация
df['purchase_log'] = np.log1p(df['purchase_amount'])  # log(x+1)

print(df)
```

```python
from sklearn.preprocessing import KBinsDiscretizer

# Групиране на purchase_amount в 3 интервала
discretizer = KBinsDiscretizer(
    n_bins=3,                # брой интервали
    encode="ordinal",        # представяне с цели числа
    strategy="uniform"       # равни интервали
)
# Преобразуване на колоната purchase_amount
df["purchase_amount_binned"] = discretizer.fit_transform(df_filled[["purchase_amount"]])
print(df)
```

```python
# IQR метод
Q1_age = df['age'].quantile(0.25)
Q3_age = df['age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

Q1_purchase = df['purchase_amount'].quantile(0.25)
Q3_purchase = df['purchase_amount'].quantile(0.75)
IQR_purchase = Q3_purchase - Q1_purchase

# Маркиране на outliers
df['age_outlier'] = ((df['age'] < Q1_age - 1.5*IQR_age) | (df['age'] > Q3_age + 1.5*IQR_age)).astype(int)
df['purchase_outlier'] = ((df['purchase_amount'] < Q1_purchase - 1.5*IQR_purchase) | 
                          (df['purchase_amount'] > Q3_purchase + 1.5*IQR_purchase)).astype(int)

df
```
