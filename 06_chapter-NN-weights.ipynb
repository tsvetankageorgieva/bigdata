```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import pandas as pd
import numpy as np

# Зареждане на данните
data = pd.read_csv('students-online.csv')
# Избор само на числовите колони
numeric_df = data.select_dtypes(include=['number'])

# Целева променлива
y = numeric_df["Satisfaction"]
# Входни характеристики
X = numeric_df.drop(columns=["Satisfaction", "ID"])
# Разделяне на данните
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)

# Създаване на невронна мрежа
model = MLPClassifier(hidden_layer_sizes=(10, 5),
    activation='relu', max_iter=1000, random_state=42)
# Обучение
model.fit(X_train, y_train)

# Извеждане на теглата
for i, weights in enumerate(model.coefs_):
    print("Тегла между слой {i} и слой {i+1}")
    print(weights)

    print("Средна абсолютна стойност:")
    print(np.mean(np.abs(weights)))
```
