```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Данни
df = pd.read_csv(r'D:\my-repo\data\students-online.csv')

# Целевият клас
y = df['Satisfaction']
X = df.drop(['ID', 'Satisfaction'], axis=1)

# Преобразуване на категорийни атрибути в числови
cat_cols = ['Gender', 'Degree', 'PlatformUsage', 'PreferredFormat']

ct = ColumnTransformer(transformers=[('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), cat_cols)],remainder='passthrough')
X_transformed = ct.fit_transform(X)

# Имена на новите колони
new_column_names = ct.get_feature_names_out()
# Обратно в DataFrame
X_final = pd.DataFrame(X_transformed, columns=new_column_names, index=X.index)
# Разделяне на данните
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, stratify=y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, precision_score, recall_score
# Decision tree модел 
dt = DecisionTreeClassifier(criterion='gini',   # или 'entropy'
                            max_depth=5)
dt.fit(X_train, y_train)

# Прогнозиране и оценка 
y_pred = dt.predict(X_test)

# Мерки
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average="macro")
print("accuracy =", accuracy)
print("f1_macro =", f1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Decision tree модел за регресия или Random Forest
dtr = DecisionTreeRegressor(criterion='squared_error', # или 'absolute_error'
                            max_depth=5)
dtr.fit(X_train, y_train)

# Прогнозиране и оценка 
y_pred = dtr.predict(X_test)

# Мерки
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("R²:", r2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, precision_score, recall_score

# Random Forest модел
rf = RandomForestClassifier(criterion="gini", 
                            n_estimators=100, max_depth=5)
rf.fit(X_train, y_train)

# Прогнозиране и оценка 
y_pred = rf.predict(X_test)

# Мерки
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
rec = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average="macro")
print("Confusion matrix:\n", cm)
print("accuracy =", accuracy)
print("Precision =", prec)
print("Recall =", rec)
print("f1_macro =", f1)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize

# вероятности за всички класове
y_scores = dt.predict_proba(X_test)

# бинаризация one-vs-rest
y_test_bin = label_binarize(y_test, classes=dt.classes_)

# ROC AUC за многокласов случай
auc_ovr = roc_auc_score(y_test_bin, y_scores, multi_class='ovr')

print("ROC AUC (OvR) =", auc_ovr)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# примерни данни
data = pd.DataFrame({
    "Temperature": [30, 32, 28, 35, 36, 31, 29, 40, 38, 27],
    "Humidity":    [70, 65, 80, 60, 58, 72, 85, 55, 57, 90],
    "Play":        [1,   1,  0,  1,  1,  0,  0,  1,  1,  0]
})

X = data[["Temperature", "Humidity"]]
y = data["Play"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
